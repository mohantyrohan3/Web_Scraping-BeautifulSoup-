# Print the title of the first 5 blogs written by Coding Ninjas

```python
import requests
from bs4 import BeautifulSoup
```

```python
r=requests.get('https://medium.com/codingninjas-blog')
r
```

```python
html=r.text
data=BeautifulSoup(html,'html.parser')
data
```

```python
# Answer
k=0
temp1=data.find_all('h3')
for i in temp1:
    if k!=5:
        print(i.get_text())
    else:
        break
    k+=1
```
